Exploratory Data Analysis (EDA) ----------------------------------------------------------------------------------

```python
#GENDER
import seaborn as sb
```

```python
df.columns
```

```python
df.head()
```

```python
gender_plot  = sb.countplot(x = 'Gender', data = df)
```

```python
# include values too:
gender_plot  = sb.countplot(x = 'Gender', data = df)

for i in gender_plot.containers:
    gender_plot.bar_label(i)
```

```python
sales = df.groupby(['Gender'], as_index=False)   # i.e. Gender_as_index = False

for gen, data in sales:
    print("\n-------------------------------------------------------------------------------------")
    print("GENDER : ",gen)
    print("\n")
    print("DATA : ",data)
```

```python
sales = df.groupby(['Gender'], as_index=False).sum().sort_values(by = 'Amount', ascending= False)

#sales
sales[['Gender','Amount']]
```

```python
# Plotting the above on bar graph
plot2 = sb.barplot(x = 'Gender', y= 'Amount', data = sales)

plot2

# for i in plot2.containers:
#     plot2.bar_label(i)
```

*Conclusion for above graph : most of the buyers are females*

**Age:**

```python
df.columns
```

```python
age_plot = sb.countplot(x = 'Age Group', data = df)
age_plot
```

```python
age_plot = sb.countplot(x = 'Age Group', data = df, hue = 'Gender')

# for count
for i in age_plot.containers:
    age_plot.bar_label(i)
```

```python
age_group = df.groupby(['Age Group'], as_index=False).sum().sort_values(by = 'Amount', ascending= False)
sb.barplot(x = 'Age Group', y = 'Amount', data = age_group)

```

*Conclusion for above age graph: Most of the buyers are of age group between 26-35 females*

**STATE**

```python
df.columns
```

```python
df.head()
```

```python
sales_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by='Orders', ascending=False)

# for state , data in sales_state:
#     print("STATE : ", state)
#     print("\n-----------------------------------------------------------------------------")
#     print("DATA: ", data)
sales_state
```

```python
sb.set(rc={'figure.figsize' : (30,9)})
sb.barplot(x='State', y = 'Orders', data = sales_state)
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sales_state
```

```python
sb.barplot(x = 'State', y = 'Amount', data = sales_state)
```

*Conclusion based on orders/amount : Most of the orders and Amount spent are from Uttar Pradesh, Maharashtra and Karnataka respectively*


**MARITAL STATUS**

```python
sb.countplot(x = 'Marital_Status', data = df)
sb.set(rc = {'figure.figsize':(6,6)})
```

```python
marital_stat = sb.countplot(x = 'Marital_Status', data = df)
sb.set(rc = {'figure.figsize':(1,1)})
for i in marital_stat.containers:
    marital_stat.bar_label(i)
```

```python
marital_group = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sb.set(rc = {'figure.figsize':(4,3)})
sb.barplot(x = 'Marital_Status', y = 'Amount', data = marital_group , hue = 'Gender')
```

*Conclusion from above graph: Most of the buyers are married women*

**OCCUPATION:**

```python
sb.set(rc = {'figure.figsize':(20,10)})
occ = sb.countplot(x = 'Occupation', data = df)

for i in occ.containers:
    occ.bar_label(i)
```

```python
occ_group = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sb.set(rc = {'figure.figsize':(20,10)})
sb.barplot(x = 'Occupation', y='Amount', data = occ_group)
```

*Conclusion from above graph: Most of the buyers are working in IT sector, HealthCare and Aviation*

```python
df.columns
```

**PRODUCT CATEGORY**

```python
product = sb.countplot(x = 'Product_Category', data = df)
for i in product.containers:
    product.bar_label(i)
```

```python
product_grp = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head()

sb.set(rc = {'figure.figsize':(10,6)})
sb.barplot(x = 'Product_Category', y = 'Amount', data = product_grp)
```

*Conclusion from above graph : Most of the sold products are from food, footwear and electronics & gadgets*

Above conclusions:


1. Most of the buyers are females
2. Most of the buyers are of age group between 26-35 females
3. Most of the orders and Amount spent are from Uttar Pradesh, Maharashtra and Karnataka respectively
4. Most of the buyers are married women
5. Most of the buyers are working in IT sector, HealthCare and Aviation
6. Most of the sold products are from food, footwear and electronics & gadgets

**Final Conclusion:**

**The dataset primarily caters to a female demographic aged between 26 and 35, residing predominantly in urban areas of Maharashtra, Uttar Pradesh, and Karnataka. These customers, often married and employed in sectors like IT, healthcare, and aviation, demonstrate a preference for products in the food, footwear, and electronics & gadgets categories.**

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

**Understanding the User**

**1. AGE DISTRIBUTION:**

```python
df.columns
```

```python
# Count the occurrences of each age value
age_counts = df['Age'].value_counts().sort_index()
#age_counts.head()
# print(age_counts.index)
# print("AGE COUNTS : ",age_counts)
```

```python

sb.set(rc={'figure.figsize':(20,4)})

age_count = sb.barplot(x=age_counts.index, y=age_counts)

# Identify the maximum value
max_value = age_counts.max()

# colors for each bar 
for container in age_count.containers:
    for bar in container:
        if bar.get_height() == max_value:
            bar.set_facecolor('red')  # Color the max bar red
        else:
            bar.set_facecolor('blue')  # Color other bars blue

# labels to the bars
for container in age_count.containers:
    age_count.bar_label(container)

plt.xlabel('Age')
plt.ylabel('Number of users')
plt.title('Age Distribution of Social Media Users')

```

*Conclusion : 
The age group of 43 years has the highest representation among social media users in the dataset.*

```python

# average Time Spent by age group:
avg_time_spent = df.groupby(['Age'])['Total Time Spent'].mean().reset_index().sort_values(by = 'Total Time Spent', ascending=False)
avg_time_spent.head()
```

```python
# Bar Plot of Average Time Spent:
sb.set(rc={'figure.figsize':(25,4)})
avg = sb.barplot(x = 'Age' , y = 'Total Time Spent', data = avg_time_spent)
plt.xlabel('Age')
plt.ylabel('Average Time Spent')
plt.title("Avg time spent on social media by Age group")

max_value =  avg_time_spent['Total Time Spent'].max()

for container in avg.containers:
    for bar in container:
        if bar.get_height() == max_value:
            bar.set_facecolor('red')
        else:
            bar.set_facecolor('blue')
    

for i in avg.containers:
    print(avg.bar_label(i))

```

*Conclusion :People in age group 37 are more prone to time-wasting, while those in age group 32 are least prone to time-wasting based on average time spent on social media.*

**2. GENDER DISTRIBUTION:**

```python
df.columns
```

```python
#  Gender Distribution :
gender_counts = df['Gender'].value_counts()
sb.set(rc={'figure.figsize':(3,4)})

# countplot to visualize the gender distribution / frequency of each gender

gen_count = sb.barplot(x=gender_counts.index, y=gender_counts)

max_value = gender_counts.max()

# setting color for max value
for container in gen_count.containers:
    for bar in container:
        if bar.get_height()==max_value:
            bar.set_facecolor('red')
        else:
            bar.set_facecolor('blue')

plt.xlabel('Gender')
plt.ylabel('Number of users')
plt.title('Gender Distribution of Social Media Users')

for i in gen_count.containers:
    gen_count.bar_label(i)
```

***Conclusion** : The dataset shows a higher number of **male users** compared to female and other gender categories.*

```python
# average Time Spent by Gender:
avg_time_spent_by_gen = df.groupby(['Gender'])['Total Time Spent'].mean().reset_index().sort_values(by = 'Total Time Spent', ascending=False)
avg_time_spent_by_gen.head()
```

```python
# Bar Plot of Average Time Spent by Gender
sb.set(rc={'figure.figsize':(4,4)})
gen_bar = sb.barplot(x = 'Gender',y = 'Total Time Spent', data = avg_time_spent_by_gen)
plt.xlabel('Gender')
plt.ylabel('Average Time Spent')
plt.title("Avg Time spent on social media by gender")


for i in gen_bar.containers:
    gen_bar.bar_label(i)


```

***Conclusion**: 
Although males are the largest group, they don't spend the most time on social media on average. Users in the **"Other"** gender category, even though they are the smallest group, spend the most time on social media.*

**3. Income Analysis:**

```python
df.columns
```

```python
df['Income'].min()
```

```python

# Grouping Income in ranges

income_range = [20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]

# Create a new column with income categories
df['Income Group'] = pd.cut(df['Income'], bins=income_range, right=False)

# Count the occurrences of each income group
income_counts = df['Income Group'].value_counts().sort_index()
print(income_counts, df['Income'])
```

```python
sb.set(rc={'figure.figsize':(15,4)})
income_count = sb.barplot(x=income_counts.index, y=income_counts)
plt.xlabel('Income Group')
plt.ylabel('Number of Users')
plt.title('Income Distribution of Users')


for i in income_count.containers:
    income_count.bar_label(i)

plt.show()
```

Analyze Average Time Spent by Income Group

```python
avg_time_spent_by_income = df.groupby(['Income Group'],observed=False)['Total Time Spent'].mean().reset_index()
 
# Bar plot for average time spent
sb.set(rc={'figure.figsize':(15,4)})
avg_time_plot = sb.barplot(x='Income Group', y='Total Time Spent', data=avg_time_spent_by_income)
plt.xlabel('Income Group')
plt.ylabel('Average Time Spent (minutes)')
plt.title('Average Time Spent on Social Media by Income Group')

# Add bar labels
for i in avg_time_plot.containers:
    avg_time_plot.bar_label(i)
```

*Conclusion : users in the 20,000-30,000 income group spend more time on social media than users in the 50,000-60,000 income group . So this suggests that income may influence how people engage with social media platforms.*


4. **Profession Analysis**:
   - Visualize the distribution of users by profession.
   - Analyze average time spent on social media by profession.**

```python
df['Profession']
```

```python
prof_counts = df['Profession'].value_counts().sort_index(ascending=False)
prof_counts

prof_count = sb.barplot(x = prof_counts.index, y = prof_counts)
plt.xlabel('Profession')
plt.ylabel('Number of users')
plt.title("Distribution of Users by Profession")
```

*Conclusion: Students use social media more than users from other professions, making them the largest group in the dataset*

```python
# average time spent on social media by profession

avg_time_spent_by_prof = df.groupby(['Profession'])['Total Time Spent'].mean().reset_index()
avg_time = sb.barplot(x = 'Profession', y='Total Time Spent',data = avg_time_spent_by_prof)
plt.xlabel('Profession')
plt.ylabel('Average Time Spent')
plt.title('Average Time Spent on Social Media by Profession')
for i in avg_time.containers:
    avg_time.bar_label(i)
```

*Conclusion :  cashiers spend the most time on social media compared to other professions.*

So, Students are the largest group of social media users, while cashiers spend the most time on social media, indicating distinct patterns of usage among different professions.

5. **Addiction Level**:
   - Visualize the distribution of addiction levels.
   - Analyze average time spent on social media by addiction level.

```python
df['Addiction Level']
```

```python
adict_count = df['Addiction Level'].value_counts().sort_index(ascending=False)
adict_count

#count plot
adict_counts = sb.barplot(x = adict_count.index, y = adict_count)
plt.xlabel("Adiction Level")
plt.ylabel("Number of Users")
plt.title("Distribution of Adiction Levels")

for i in adict_counts.containers:
    adict_counts.bar_label(i)
```

Conclusion : The most common addiction level among users is level 2


```python
# Average time spent on social media by addiction level
avg_time_spent_adict = df.groupby(['Addiction Level'])['Total Time Spent'].mean().reset_index()
avg_time = sb.barplot(x = 'Addiction Level', y = 'Total Time Spent', data = avg_time_spent_adict)

plt.ylabel("Average Time Spent on Adiction Level")


for i in avg_time.containers:
    avg_time.bar_label(i)
```

Conclusion : Users with an addiction level of 7 spend the most time on social media compared to other addiction levels.




**The EDA reveals key insights regarding users on social media**:

- **Age**: Users aged 37 tend to waste the most time on social media.
- **Gender**: While students make up the largest group of social media users, those in the "Other" gender category spend the most time online.
- **Profession**: Cashiers, despite being fewer in number, spend the most time on social media compared to other professions.
- **Income**: Users in the 20,000-30,000 income range engage more with social media than users in higher income brackets.
- **Addiction Level**: The most common addiction level among users is level 2, whereas users with an addiction level of 7 spend the most time on social media.

Overall, these findings illustrate how various demographic factors influence social media usage and highlight potential concerns regarding time-wasting behaviors.
